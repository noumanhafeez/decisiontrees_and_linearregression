```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("../data/mushrooms.csv")

# Initialize encoder
le = LabelEncoder()

# Convert all columns to integers
for col in df.columns:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
# Compute the correlation matrix
corr = df.corr()  # Pearson correlation

# Plot the correlation matrix
plt.figure(figsize=(12, 10))
plt.matshow(corr, fignum=1, cmap='coolwarm')  # Matshow inside figure
plt.colorbar()  # Show color scale
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix of Mushroom Dataset", pad=30)
plt.show()
```

```python
process_data = df[['class', 'cap-surface', 'cap-color', 'odor', 'gill-size', 'population', 'habitat']]
```
